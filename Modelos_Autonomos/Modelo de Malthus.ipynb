```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

### Modelo de Dinâmica Populacional
- Modelo de Malthus

Seja $P(t)$ a população de uma dada espécie no instante $t$. A hipótese de Malthus
sobre a variação da população é que a taxa de variação de y é proporcional ao valor de $P$,
ou seja,

$$\dfrac{dP}{dt} = rP $$

onde $r$ é a denominada a taxa de crescimento ou declínio, dependendo se é positiva ou
negativa, respectivamente.

Vamos supor que $r > 0$, de modo que a população está crescendo. Sob a condição
inicial $P(0) = P_0$ , temos o seguinte PVI:

$$\begin{cases} 
\dfrac{dP}{dt} = rP \\ \\
P(0) = P_0
\end{cases}$$

```python
# Definindo a EDO
r = 0.006
dydt = lambda t,P: r*P

# Vetor de tempo
t = np.linspace(0,100,100)

# Definindo a condição Inicial
P0 = 5

# Resolver Numericamente
sol = solve_ivp(dydt,[0,100],[P0],t_eval=t)
```

```python
# Visualizar graficamente
plt.style.use("dark_background")
fig,ax = plt.subplots()
ax.plot(sol.t,sol.y[0])
ax.set_xlabel("tempo t")
ax.set_ylabel("y(t)")
ax.set_title("Solução Númerica")
```
